+ Regression models the relationship between dependednt and independent variables
+ how much y changes when x changes
+ Unlike clasification , regression pridicts continuous values
+ regression is about trends, not exact predections

+ Regression is simple supervised ML algorithm
+ helps in prediction,estimation and decision making
+ widely used in economics, management,engineering, health and ML
+ Forms the foundation of many advanced ML algorithm 

Types of Regression
+ Simple Linear Regression --> one input,one output
+ Multiple Linear Regression --> multiple inputs
+ Polynomial Regression --> captures non linear patterns
+ Logistic Regression --> classification (despite the name)

Regression equation
+ β0 --> intercept
+ β1 --> slope(impact of x on y)
+ ε --> error term(unexplained variation)
  
  Y = β0 + β1X + ε

Assumption of linear Regression
+ Linearity
+ independence error
+ constant variance
+ Normality of residuals
+ No multicollinearity(for multiple regression)
+ if assumptions fail, results become unreliable

How Regression learn
+ Minimises error between actual and predicted values
+ Uses Latest Squares Method
+ Best fit line minimizes total squared error

Model Evaluation mrtrics
+ MSE/RMSE
